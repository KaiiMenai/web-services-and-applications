Importing the modules needed for requesting the URL and for counting the ouput

```python
import requests
from collections import Counter # for future reference, this is case sensitive - had errors when using lowercase.
```

## PART 1

### STEP 1
#### Shuffle a new deck

Outline logic

Send a GET request. 
I start by calling the shuffle endpoint provided by the API.

Convert response to JSON
This returns a deck_id, which uniquely identifies the deck.


After getting the program to read deck_id, I print the deck ID which is good for debugging, should errors arise. The deck_id is saved so that the same deck is used when drawing cards.

```python

shuffle_url = "https://deckofcardsapi.com/api/deck/new/shuffle/?deck_count=1"
response = requests.get(shuffle_url)
data = response.json()

deck_id = data["deck_id"]
print("Deck ID:", deck_id)
```

## PART 2

### STEP 2
#### Draw 5 cards from the shuffled deck

Here I want to use the saved deck_id to draw 5 cards in the same request.

After rebuilding the URL from the example with the deck_id, the request will be sent, and the "cards" will be extracted from the response and stored as "hand", these will then be the cards that were drawn for the hand.
Using the '''deck_id''' from the previous step, I request 5 cards.
The API returns a list of card objects, each containing information such as "value" and "suit".

```python
draw_url = f"https://deckofcardsapi.com/api/deck/{deck_id}/draw/?count=5" # deck_id is inserted into the url using an f-stirng, this is to ensure that the same deck is used for drawing the cards, and the count parameter specidfies the number of cards to draw.
response = requests.get(draw_url)
data = response.json()

hand = data["cards"]
```

## Part 3

### STEP 3
#### Print the cards drawn

At this stage, the output needs to be printed out in a readable format to aid in interpretation, so need to convert the raw API.
I used a loop that prints each card in a readable format. Only the value and suit are required for the assignment, so those are extracted from each card dictionary.


```python
print("\nYour hand:")
for card in hand:
    value = card["value"]
    suit = card["suit"]
    print(f"{value} of {suit}")
```

## PART 4

### STEP 4
#### Extract values and suits into separate lists

Making sure the data is clean to make hand analysis easier. For this getting a list of card values and card suits is needed. This is then looped throught the hand.

To make checking for pairs, straights, and flushes easier, I store all card values in one list and all suits in another. This simplifies counting and comparison later on.

```python
values = []
suits = []

for card in hand:
    values.append(card["value"])
    suits.append(card["suit"])
```

## Part 5

Checking for combinations in the output.

### STEP 5
#### Check for pairs and three-of-a-kind

A. Pair / Three-of-a-kind

Count how many times each value appears

Use Counter(values). I use Counter to count how many times each card value appears.


Logic:

If any value appears twice → pair

If any value appears three times → three-of-a-kind.

```python
value_counts = Counter(values)

has_pair = False
has_triple = False

for count in value_counts.values():
    if count == 2:
        has_pair = True
    elif count == 3:
        has_triple = True
```

### STEP 6
#### Check for a straight

A straight requires card values to be consecutive.

Since face cards are strings, I convert each card value to a number. (ACE = 14, KING = 13, QUEEN = 12, etc.)
After converting, I sort the values and check whether each number increases by exactly 1 each time.

```python
value_map = {
    "ACE": 14,
    "KING": 13,
    "QUEEN": 12,
    "JACK": 11,
    "10": 10,
    "9": 9,
    "8": 8,
    "7": 7,
    "6": 6,
    "5": 5,
    "4": 4,
    "3": 3,
    "2": 2
}

numeric_values = [value_map[v] for v in values]
numeric_values.sort()

is_straight = True
for i in range(len(numeric_values) - 1):
    if numeric_values[i + 1] != numeric_values[i] + 1:
        is_straight = False
        break
```

### STEP 7
#### Check for a flush (all cards same suit)

A flush occurs when all cards share the same suit. 
Converting the suits list to a set removes duplicates.
If the set length is 1, all suits are the same.
Very simple check.

```python
is_flush = len(set(suits)) == 1
```

### STEP 8
#### Display results and congratulate the user

Based on the checks above, the program prints messages congratulating the user if they have a special hand.

```python
print("\nResults:")

if has_triple:
    print("🎉 Congratulations! You have three of a kind!")
elif has_pair:
    print("🎉 Congratulations! You have a pair!")

if is_straight:
    print("🔥 Nice! You have a straight!")

if is_flush:
    print("💎 Amazing! All cards are the same suit (flush)!")

if not (has_pair or has_triple or is_straight or is_flush):
    print("No special hand this time — try again!")
```
